# Mengatasi Missing Value dengan Drop Data

## Menghapus Data yang Kosong (NaN)

```python
df_cleaned = df.dropna()

df_cleaned
```

## Mengatasi missing value dengan data baru

```python
df_New = df.copy()

df_New['Umur'] = df_New['Umur'].fillna(50)

df_New['Gaji'] = df_New['Gaji'].fillna(25000000)

df_New
```

## mengatasi missing value dengan nilai mean

\

```python
df_New = df.copy()

df_New['Umur'] = df_New['Umur'].fillna(df_New['Umur'].mean())

df_New['Gaji'] = df_New['Gaji'].fillna(df_New['Gaji'].mean())

df_New
```

```python
df_New = df.copy()

df_New['Umur'] = df_New['Umur'].fillna(round(df_New['Umur'].mean()))

df_New['Gaji'] = df_New['Gaji'].fillna(round(df_New['Gaji'].mean()))

df_New
```

## Mengatasi missing value dengan Median

```python
df_New = df.copy()

df_New['Umur'] = df_New['Umur'].fillna(round(df_New['Umur'].median()))

df_New['Gaji'] = df_New['Gaji'].fillna(round(df_New['Gaji'].median()))

df_New
```

## Mengatasi missing value dengan mode

```python
df_New = df.copy()

df_New['Umur'] = df_New['Umur'].fillna(round(df_New['Umur'].mode()[0]))

df_New['Gaji'] = df_New['Gaji'].fillna(round(df_New['Gaji'].mode()[2]))

df_New
```

## mengatasi wrong data dengan ubah data

```python
df_New = df.copy()

df_New.loc[df['Umur']>120, 'Umur'] = 33

df_New
```

# Soal 4

```python
import pandas as pd

# Load the data
df_customers = pd.read_csv('/content/customers.csv')  # File data pelanggan
df_gasstations = pd.read_csv('/content/gasstations.csv')  # File data SPBU
df_products = pd.read_csv('/content/products.csv')  # File data produk
df_transactions = pd.read_csv('/content/transactions_1k.csv')  # File data transaksi

# Merge the dataframes
df_combined = pd.merge(df_transactions, df_customers, on='CustomerID', how='left')
df_combined = pd.merge(df_combined, df_gasstations, on='GasStationID', how='left')
df_combined = pd.merge(df_combined, df_products, on='ProductID', how='left')

# Convert 'Date' column to datetime format
df_combined['Date'] = pd.to_datetime(df_combined['Date'])

# Filter data for specific dates (e.g., 2012-08-23 to 2012-08-26)
filtered_dates = ['2012-08-23', '2012-08-24', '2012-08-25', '2012-08-26']
df_filtered = df_combined[df_combined['Date'].isin(pd.to_datetime(filtered_dates))]

# Group by 'Date' and calculate descriptive statistics
df_result = df_filtered.groupby('Date')[['Amount']].describe()

df_result.head()
```

```python
df_transactions['Date'] = pd.to_datetime(df_transactions['Date'])
df_transactions['Day'] = df_transactions['Date'].dt.day

filtered_df = df_transactions[df_transactions['Day'].isin([23, 24, 25, 26])]
stats_by_day = filtered_df.groupby('Day')[['Amount', 'Price']].describe()

stats_by_day
```

# Praktikum 5

```python
# Mengubah Time menjadi periode per jam
df_combined['Hour'] = pd.to_datetime(df_combined['Time'], format='%H:%M:%S').dt.hour
# Mengelompokkan berdasarkan tanggal dan jam
df_group = df_combined.groupby(['Date', 'Hour'])[['Amount', 'Price']].sum()
df_group.rename(columns={'Price': 'Gas Sales'}, inplace=True)
df_group
```

```python
best_time = df_group.reset_index().sort_values(['Date', 'Amount'], ascending=[True, False])
best_hour_day = best_time.groupby('Date').first()
best_hour_day.reset_index(inplace=True)
best_hour_day
```
